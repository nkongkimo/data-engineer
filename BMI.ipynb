```python
# @title
import sys # 導入 sys 模組，用於系統相關功能，例如 sys.exit() 終止程式

# bmi: weight(kg) / height(m) ^ 2 # 這行是註解，說明 BMI 的計算公式：體重(公斤) / (身高(公尺))^2

# Input for weight with validation # 體重輸入及驗證的區塊開始
while True: # 啟動一個無限循環，直到使用者輸入有效數值為止
    try: # 嘗試執行以下程式碼
        weight_input = input("請輸入體重(公斤): ") # 提示使用者輸入體重，並將輸入儲存為字串
        weight = float(weight_input) # 嘗試將輸入的字串轉換為浮點數（小數）
        break # 如果轉換成功，則跳出 while 循環
    except ValueError: # 如果轉換失敗（例如輸入了非數字），則捕獲 ValueError 錯誤
        print("輸入錯誤：請輸入數字作為體重。") # 提示使用者輸入錯誤，需輸入數字

# Input for height with validation # 身高輸入及驗證的區塊開始
while True: # 啟動一個無限循環，直到使用者輸入有效數值為止
    try: # 嘗試執行以下程式碼
        height_input = input("請輸入身高(公分): ") # 提示使用者輸入身高，並將輸入儲存為字串
        height = float(height_input) # 嘗試將輸入的字串轉換為浮點數（小數）
        break # 如果轉換成功，則跳出 while 循環
    except ValueError: # 如果轉換失敗，則捕獲 ValueError 錯誤
        print("輸入錯誤：請輸入數字作為身高。") # 提示使用者輸入錯誤，需輸入數字

if height <= 0 or weight <= 0: # 檢查身高或體重是否小於或等於零
    print("體重和身高必須是正數。") # 如果是，則印出錯誤訊息
    sys.exit() # 終止程式執行

bmi = weight / (height / 100) ** 2  # 計算 BMI 值，身高(公分)需先除以 100 轉換為百分比
print(f"你的體重是: {weight} 公斤")  # 使用 f-string 格式化輸出體重
print(f"你的身高是: {height} 公分")  # 使用 f-string 格式化輸出身高
print(f"你的 BMI 是: {bmi:.2f}")    # 使用 f-string 格式化輸出 BMI，並將結果格式化為兩位小數

# BMI 判斷式
if bmi >= 25:
    print("體重過重")
elif bmi >= 18.5:
    print("體重正常")
else:
    print("體重過輕")


```
